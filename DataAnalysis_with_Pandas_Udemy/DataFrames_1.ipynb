## Reshaping, sorting, transposing

```python
df = pd.DataFrame({
    'col1': [2, 100, 300, np.nan, 400, 250],
    'col2': [4, 1, 9, 2, 7, np.nan],
    'col3': [6, 1.0, 9, 9, 2, 3],
})
df
```

```python
df.sort_values(by=["col1","col2"], axis = 0)
df.sort_values(by=1, axis = 1)
```

```python
df
df.sum(axis = 0,skipna=False)
df.kurt(axis=0)
df.mode(axis = 0)
df.mode(axis = 1)
df.cummax(axis = 1, skipna=False)
df.sem(axis = 0)
df.sem(axis = 1)
```

## Explode method

```python
df = pd.DataFrame({'A': [[1, 2, 3], 'foo', [], [3, 4]], 'B': [[100,200],[20,30],40,50]})
df
```

```python
df_exploded = df.explode(column = 'A').explode('B')
```

```python
df_exploded.sort_index(axis = 1, ascending=False)
```

## Binary Operations

```python
df = pd.DataFrame({
    'col1': [2, 100, 300, np.nan, 400, 250],
    'col2': [4, 1, 9, 2, 7, np.nan],
    'col3': [6, 1.0, 9, 9, 2, 3]})
df_2 = pd.DataFrame({
    'col1': [-2, -100, -300, -10000, -400, -250],
    'coly': [-4, -1, -9, -2, -7, -100000],
    'colz': [-6, -1.0, -9, -9, -2, -3]},
    index=[10,11,12,13,14,15])
s = pd.Series(data=[3,30,300], index=["col1","col2","col4"])
df
df.add(s, axis = 'columns')
df.add(df_2, axis = 0)
```

## Rank Method

```python
df = pd.DataFrame(data={'Animal': ['cat', 'penguin', 'dog',
                                   'spider', 'snake'],
                        'Number_legs': [4, 4, np.nan, 8, np.nan],
                       'Number of fingers': [20,10,20,40,60]})
df
```

```python
df.rank(axis =0, numeric_only= False, method='min')
df.rank(axis =0, numeric_only= False, na_option='bottom', method='dense')
```

```python
nba["Salary"] = nba["Salary"].fillna(0).astype("int")
nba["Salary Rank"] = nba["Salary"].rank(ascending=False).astype("int")
```

```python
nba.sort_values(by = "Salary", ascending=False)
```
